```python
import matplotlib.pyplot as plt
import numpy as np

# 2D vector [3, 2]
v = np.array([3, 2])

# Plot karte hain
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', width=0.01)
plt.xlim(0, 4)
plt.ylim(0, 3)
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector [3, 2]')
plt.show()
```

#### Matrix Multiplication (सबसे Important! ⭐⭐⭐)
Ye Linear Algebra ka heart hai! Neural Networks mein har jagah use hota hai.

Rule याद रखो:
(m × n) matrix × (n × p) matrix = (m × p) matrix

First matrix ke columns = Second matrix ke rows hone chahiye!
Example 1: Simple Matrix Multiplication

```python
# Simple matrix Multiplication

# A is 2x3
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# B is 3x2
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Result will be 2x2
C = np.dot(A, B)  # Ya A @ B
print(C)
# [[ 58  64]
#  [139 154]]
```

```python
def matrix_multiply_visual(A, B):
    """Matrix multiplication ko step by step dikhata hai"""
    m, n = A.shape
    n2, p = B.shape
    
    if n != n2:
        print("ERROR: Multiplication possible nahi hai!")
        return None
    
    C = np.zeros((m, p))
    
    print(f"Matrix A ({m}x{n}) × Matrix B ({n}x{p}) = Matrix C ({m}x{p})\n")
    
    for i in range(m):
        for j in range(p):
            total = 0
            calculation = ""
            for k in range(n):
                total += A[i,k] * B[k,j]
                calculation += f"({A[i,k]}×{B[k,j]}) + "
            
            C[i,j] = total
            print(f"C[{i},{j}] = {calculation[:-3]} = {total}")
    
    return C

# Test karo
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = matrix_multiply_visual(A, B)
print("\nFinal Result:")
print(result)
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Property 1: Matrix multiplication is NOT commutative
# A×B ≠ B×A
print("A × B:")
print(np.dot(A, B))

print("\nB × A:")
print(np.dot(B, A))
# Dekho! Different results!

# Property 2: Identity matrix
I = np.eye(2)
print("\nA × I:")
print(np.dot(A, I))
# Output: A hi milega!
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Original A:")
print(A)
# [[1 2 3]
#  [4 5 6]]

A_T = A.T  # Ya np.transpose(A)
print("\nTranspose A^T:")
print(A_T)
# [[1 4]
#  [2 5]
#  [3 6]]

print(f"\nOriginal shape: {A.shape}")      # (2, 3)
print(f"Transpose shape: {A_T.shape}")     # (3, 2)
```

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Transpose ka size reverse hoga (3,2)
A_T = np.zeros((3, 2))
for i in range(len(A)):          # rows
    for j in range(len(A[0])):   # columns
        A_T[j][i] = A[i][j]

print("\nManual Transpose:")
print(A_T)
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Property 1: (A^T)^T = A
print(np.array_equal(A, A.T.T))  # True

# Property 2: (A + B)^T = A^T + B^T
print(np.array_equal((A + B).T, A.T + B.T))  # True

# Property 3: (A × B)^T = B^T × A^T (order reverse!)
print(np.array_equal(np.dot(A, B).T, np.dot(B.T, A.T)))  # True
```

```python
A = np.array([[4, 7],
              [2, 6]])

# Inverse nikalo
A_inv = np.linalg.inv(A)
print("A inverse:")
print(A_inv)

# Verify: A × A⁻¹ = I
I = np.dot(A, A_inv)
print("\nA × A⁻¹:")
print(I)
# Almost identity (floating point errors ki wajah se thoda difference)
print(np.allclose(I, np.eye(2)))  # True
```

```python
A = np.array([[4, 7],
              [2, 6]])

det = np.linalg.det(A)
print(f"Determinant: {det}")
# Output: 10.0

# Agar determinant = 0, to inverse nahi hai
singular = np.array([[1, 2],
                     [2, 4]])
print(f"Singular matrix determinant: {np.linalg.det(singular)}")
# Output: 0.0
```

```python
# Linear Regression using matrices!
# y = X × w (find best w)

# Data
X = np.array([[1, 1],   # [bias, feature]
              [1, 2],
              [1, 3],
              [1, 4]])

y = np.array([[2],
              [4],
              [6],
              [8]])

# Formula: w = (X^T × X)^-1 × X^T × y
X_T = X.T
w = np.linalg.inv(X_T @ X) @ X_T @ y

print("Weights:")
print(w)
# [[0]    <- bias
#  [2]]   <- slope

# Prediction
y_pred = X @ w
print("\nPredictions:")
print(y_pred)

# Ye ekdum linear line fit kar diya: y = 0 + 2x
```

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

r = len(A)
c = len(B[0])

res = np.zeros((r, c))

for i in range(r):              # rows of A
    for j in range(c):          # columns of B
        for k in range(len(B)): # columns of A / rows of B
            res[i][j] += A[i][k] * B[k][j]

print(res)
```
